```python
def factorial(n):
    if n == 1:
        return 1
   return n * factorial(n-1)
```

```python
import stdio
def moves(n, left):
    if n == 0:
        return
    moves(n-1, not left)
    if left:
        stdio.writeln(str(n) + ' left')
    else:
        stdio.writeln(str(n) + ' right')
    moves(n-1, not left)
    
n = 5
moves(n, True)
```
