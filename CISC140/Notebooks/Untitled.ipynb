```python
class Animal:
    
    def __init__(self, name):
        self._name = name
        
    def name(self):
        return self._name
    
    def make_noise(self):
        return "animal noise"
```

```python
a = Animal("animal")

print(a.name() + " says " + a.make_noise())
```

```python
class Turtle(Animal):
    def __init__(self, name, size):
        super().__init__(name)
        self._size = size
        
    def size(self):
        return self._size
        
    def make_noise(self):
        return "turtle noise"
        
```

```python
t = Turtle("snapper", "large")

print(t.name() + " says " + t.make_noise())
print(t.size())
```

```python
class Person(Animal):
    def __init__(self, name, gender):
        super().__init__(name)
        self._gender = gender
        
    def make_noise(self):
        return "person noise"
```

```python
class Ninja(Person):
    def __init__(self, name, gender):
        super().__init__(name, gender)
        
    def make_noise(self):
        return " "
```

```python
class NinjaTurtle(Ninja, Turtle):
    def __init__(self, name, gender, size):
        Turtle.__init__(self, name, size)
        self._gender = gender
        
    def make_noise(self):
        return "cowabunga"
    
```

```python
nt = NinjaTurtle("Michaelangelo", "male", "medium")

print(nt._name)
print(nt._size)
print(nt._gender)
```
