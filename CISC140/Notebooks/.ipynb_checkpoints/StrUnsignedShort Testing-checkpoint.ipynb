```python
from strunsignedshort import StrUnsignedShort

print("Testing the StrUnsignedShort class.")
print()
print("1. Testing construction.")

print("   Valid values")
print("     num1 = StrUnsignedShort('four')")
num1 = StrUnsignedShort("four")
print("     num2 = StrUnsignedShort('two')")
num2 = StrUnsignedShort("two")
print("     num3 = StrUnsignedShort('one hundred and twenty-seven')") 
num3 = StrUnsignedShort("one hundred and twenty-seven")
print("     num4 = StrUnsignedShort('fifty eight')") 
num4 = StrUnsignedShort('fifty eight')
print("     num5 = StrUnsignedShort('seventeen')") 
num5 = StrUnsignedShort('seventeen')
print("     num6 = StrUnsignedShort('two hundred fifty')") 
num6 = StrUnsignedShort('two hundred fifty')

print()
print("   Invalid values")
print("     StrUnsignedShort('six hundred')")
try:
    StrUnsignedShort("six hundred")
except Exception as e:
    print("       Exception caught:", e)
    
print("     StrUnsignedShort('googelplex')")
try:
    StrUnsignedShort("googelplex")
except Exception as e:
    print("       Exception caught:", e)

print()
print("2. Testing comparison.") 
print("  {0}: {1}".format("num1 == num2", num1 == num2))
print("  {0}: {1}".format("num1 != num2", num1 != num2))
print("  {0}: {1}".format("num1 <  num2", num1 < num2))
print("  {0}: {1}".format("num1 <= num2", num1 <= num2))
print("  {0}: {1}".format("num1 >  num2", num1 > num2))
print("  {0}: {1}".format("num1 >= num2", num1 >= num2))
    
print()
print("3. Testing math operators.")     
print("  {0}: {1}".format("num1 + num2  ", num1 + num2))
print("  {0}: {1}".format("num1 + 1     ", num1 + 1))
print("  {0}: {1}".format("1 + num1     ", 1 + num1))
print("  {0}: {1}".format("num1 + 'one' ", num1 + 'one'))
print("  {0}: {1}".format("num1 - num2  ", num1 - num2))
print("  {0}: {1}".format("num1 - 1     ", num1 - 1))
print("  {0}: {1}".format("1 - num1     ", 1 - num1))
print("  {0}: {1}".format("num1 - 'one' ", num1 - 'one'))
print("  {0}: {1}".format("num1 * num2  ", num1 * num2))
print("  {0}: {1}".format("num1 * 2     ", num1 * 2))
print("  {0}: {1}".format("2 * num1     ", 2 * num1))
print("  {0}: {1}".format("num1 * 'two' ", num1 * 'two'))    
print("  {0}: {1}".format("num3 / num2  ", num3 / num2))
print("  {0}: {1}".format("num2 / 2     ", num2 / 2))
print("  {0}: {1}".format("2 / num1     ", 2 / num1))
print("  {0}: {1}".format("num1 / 'two' ", num1 / 'two'))        
print("  {0}: {1}".format("num3 // num2  ", num3 // num2))
print("  {0}: {1}".format("num2 // 2     ", num2 // 2))
print("  {0}: {1}".format("2 // num1     ", 2 // num1))
print("  {0}: {1}".format("num1 // 'two' ", num1 // 'two'))        

print()
print("4. Testing string operators.")  

print("  {0}: {1}".format("str(num1) ", str(num1)))
print("  {0}: {1}".format("str(num2) ", str(num2)))
print("  {0}: {1}".format("str(num3) ", str(num3)))
print("  {0}: {1}".format("str(num4) ", str(num4)))
print("  {0}: {1}".format("str(num5) ", str(num5)))
print("  {0}: {1}".format("str(num6) ", str(num6)))
print("  {0}: {1}".format("repr(num1)", repr(num1)))
print("  {0}: {1}".format("repr(num2)", repr(num2)))
print("  {0}: {1}".format("repr(num3)", repr(num3)))
print("  {0}: {1}".format("repr(num4)", repr(num4)))
print("  {0}: {1}".format("repr(num5)", repr(num5)))
print("  {0}: {1}".format("repr(num6)", repr(num6)))

print()
print("5. Testing the bonus operators.")
print("  {0}: {1}".format("num3 % num4  ", num3 % num4))
print("  {0}: {1}".format("num3 % 10    ", num3 % 10))
print("  {0}: {1}".format("15 % num1    ", 15 % num1))
print("  {0}: {1}".format("num3 % 'five'", num3 % 'five'))
print("  {0}: {1}".format("num1 ** num2 ", num1 ** num2))
print("  {0}: {1}".format("num1 ** 2    ", num1 ** 2))
print("  {0}: {1}".format("2 ** num1    ", 2 ** num1))
print("  {0}: {1}".format("num1 ** 'two'", num1 ** 'two')) 
print("  {0}: {1}".format("int(num5)    ", int(num5)))
print("  {0}: {1}".format("float(num6)  ", float(num6)))
```
